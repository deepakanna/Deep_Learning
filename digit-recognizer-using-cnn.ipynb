```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train=pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
df_train.head()
```

```python
j=0
for i in range(9):
    img=df_train.iloc[i,1:].values.reshape(28,28)
    plt.imshow(img)
    
    plt.show()
```

```python
df1=df_train.drop('label',axis=1)
```

```python
df2=pd.get_dummies(df_train['label'])
```

```python
df=df1.join(df2)
df.head()
```

```python
X=df1
X=X/255.0
X=X.values.reshape(-1,28,28,1)
y=df2
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True)

```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=tf.keras.Sequential([
   
    layers.Conv2D(16,(3,3),activation='relu',input_shape=(28,28,1),padding='same'),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(32,(3,3),activation='relu',padding='same'),
    layers.MaxPool2D((2,2)),
    
    layers.BatchNormalization(),
    layers.Flatten(),
    layers.Dropout(0.3),
    layers.Dense(128,activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(512,activation='relu'),
    layers.Dense(10,activation='softmax')
    
    
])
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(X_train,y_train,epochs=15,validation_data=(X_test,y_test),batch_size=32)
```

```python
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']
epoch_range=range(15)
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
plt.plot(epoch_range,acc,label='Training Accuracy')
plt.plot(epoch_range,val_acc,label='Validation Accuracy')
#plt.ylim(0.8,1)
plt.subplot(1,2,2)
plt.plot(epoch_range,loss,label='Training loss')
plt.plot(epoch_range,val_loss,label='Validation loss')
#plt.ylim(0.02,0.08)

plt.show()
```

```python
test_df=pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
test_df.head()
```

```python
test_df1=test_df/255.
test_df1=test_df1.values.reshape(-1,28,28,1)
```

```python
pred=model.predict(test_df1)

```

```python
pred_labels=np.argmax(pred,axis=1)
pred_labels
```
